## data_loading

```python
import pandas as pd
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
features = pd.read_csv("features.csv")
stores = pd.read_csv("stores.csv")

print("Train Shape :", train.shape)
print("Test Shape :", test.shape)
print("Features Shape :", features.shape)
print("Stores Shape :", stores.shape)

# Merge train dataset
train_df = pd.merge(
    train,
    features,
    on=["Store", "Date", "IsHoliday"],
    how="left"
)

train_df = pd.merge(
    train_df,
    stores,
    on="Store",
    how="left"
)

# Merge test dataset
test_df = pd.merge(
    test,
    features,
    on=["Store", "Date", "IsHoliday"],
    how="left"
)

test_df = pd.merge(
    test_df,
    stores,
    on="Store",
    how="left"
)

print("\nMerged Train Shape :", train_df.shape)
print("Merged Test Shape :", test_df.shape)

print("\nTrain Data")
print(train_df.head())

print("\nTest Data")
print(test_df.head())

# Save merged files
train_df.to_csv("merged_train.csv", index=False)
test_df.to_csv("merged_test.csv", index=False)

print("\nMerged datasets saved successfully.")
```

## preprocessing

```python
# ==========================================
# Retail Demand Forecasting
# Data Preprocessing
# ==========================================

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load merged datasets
train = pd.read_csv("merged_train.csv")
test = pd.read_csv("merged_test.csv")

print("Train Shape :", train.shape)
print("Test Shape :", test.shape)

# ------------------------------------------
# Missing Values
# ------------------------------------------

print("\nMissing Values Before Cleaning")
print(train.isnull().sum())

markdown_cols = [
    "MarkDown1",
    "MarkDown2",
    "MarkDown3",
    "MarkDown4",
    "MarkDown5"
]

# Fill MarkDown columns with 0
for col in markdown_cols:
    train[col] = train[col].fillna(0)
    test[col] = test[col].fillna(0)

# Fill CPI and Unemployment using Median
train["CPI"] = train["CPI"].fillna(train["CPI"].median())
test["CPI"] = test["CPI"].fillna(test["CPI"].median())

train["Unemployment"] = train["Unemployment"].fillna(train["Unemployment"].median())
test["Unemployment"] = test["Unemployment"].fillna(test["Unemployment"].median())

print("\nMissing Values After Cleaning")
print(train.isnull().sum())



train.drop_duplicates(inplace=True)
test.drop_duplicates(inplace=True)

print("\nTrain Shape After Removing Duplicates :", train.shape)
print("Test Shape After Removing Duplicates :", test.shape)

# ------------------------------------------
# Convert Date
# ------------------------------------------

train["Date"] = pd.to_datetime(train["Date"])
test["Date"] = pd.to_datetime(test["Date"])

# ------------------------------------------
# Date Features
# ------------------------------------------

for df in [train, test]:

    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
    df["Day"] = df["Date"].dt.day
    df["Quarter"] = df["Date"].dt.quarter

encoder = LabelEncoder()
train["Type"] = encoder.fit_transform(train["Type"])
test["Type"] = encoder.transform(test["Type"])
train["IsHoliday"] = train["IsHoliday"].astype(int)
test["IsHoliday"] = test["IsHoliday"].astype(int)
print("\nData Types")
print(train.dtypes)
train.to_csv("clean_train.csv", index=False)
test.to_csv("clean_test.csv", index=False)

print("\nData preprocessing completed successfully.")
print("Files Saved:")
print("clean_train.csv")
print("clean_test.csv")
```

## Feature Engineering

```python
import pandas as pd

train = pd.read_csv("clean_train.csv")
test = pd.read_csv("clean_test.csv")
train["Date"] = pd.to_datetime(train["Date"])
test["Date"] = pd.to_datetime(test["Date"])
print("Creating Date Features...")

for df in [train, test]:

    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
    df["Quarter"] = df["Date"].dt.quarter

for df in [train, test]:
    df["Total_Markdown"] = (
        df["MarkDown1"] +
        df["MarkDown2"] +
        df["MarkDown3"] +
        df["MarkDown4"] +
        df["MarkDown5"]
    )
train.drop(
    ["Date", "MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"],
    axis=1,
    inplace=True
)

test.drop(
    ["Date", "MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"],
    axis=1,
    inplace=True
)
train.to_csv("feature_train.csv", index=False)
test.to_csv("feature_test.csv", index=False)
```

## model_training

```python
import pandas as pd
import joblib

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ------------------------------------------
# Load Dataset
# ------------------------------------------

df = pd.read_csv("feature_train.csv")

print("Dataset Shape :", df.shape)

# ------------------------------------------
# Features and Target
# ------------------------------------------

X = df.drop("Weekly_Sales", axis=1)
y = df["Weekly_Sales"]

# ------------------------------------------
# Time-Based Train-Test Split
# ------------------------------------------

split = int(len(df) * 0.80)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]

print("Training Shape :", X_train.shape)
print("Testing Shape  :", X_test.shape)

# ==========================================
# Linear Regression
# ==========================================

print("\nTraining Linear Regression...")

lr = LinearRegression()

lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)

# ==========================================
# Random Forest
# ==========================================

print("\nTraining Random Forest...")

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

# ==========================================
# XGBoost
# ==========================================

print("\nTraining XGBoost...")

xgb = XGBRegressor(
    objective="reg:squarederror",
    random_state=42
)

xgb.fit(X_train, y_train)

# ==========================================
# Hyperparameter Tuning
# ==========================================

print("\nStarting Hyperparameter Tuning...")

params = {

    "n_estimators": [100, 200],

    "max_depth": [4, 6, 8],

    "learning_rate": [0.05, 0.1],

    "subsample": [0.8, 1.0],

    "colsample_bytree": [0.8, 1.0]

}

grid = GridSearchCV(

    estimator=XGBRegressor(
        objective="reg:squarederror",
        random_state=42
    ),
    param_grid=params,

    scoring="neg_mean_absolute_error",

    cv=3,

    n_jobs=-1,

    verbose=2

)

grid.fit(X_train, y_train)

print("\nBest Parameters:")
print(grid.best_params_)

best_model = grid.best_estimator_

# ==========================================
# Predictions
# ==========================================

pred_xgb = best_model.predict(X_test)

# ==========================================
# Evaluation Function
# ==========================================

def evaluate_model(name, actual, prediction):

    mae = mean_absolute_error(actual, prediction)
    mse = mean_squared_error(actual, prediction)
    rmse = mse ** 0.5
    r2 = r2_score(actual, prediction)

    print("\n" + "="*40)
    print(name)
    print("="*40)

    print("MAE       :", round(mae, 2))
    print("MSE       :", round(mse, 2))
    print("RMSE      :", round(rmse, 2))
    print("R2 Score  :", round(r2, 4))

# ==========================================
# Evaluate Models
# ==========================================

evaluate_model("Linear Regression", y_test, pred_lr)

evaluate_model("Random Forest", y_test, pred_rf)

evaluate_model("XGBoost (Tuned)", y_test, pred_xgb)

# ==========================================
# Save Models
# ==========================================

joblib.dump(lr, "linear_regression.pkl")

joblib.dump(rf, "random_forest.pkl")

joblib.dump(best_model, "xgboost_best.pkl")

print("\nModels Saved Successfully.")

# ==========================================
# Save Predictions
# ==========================================

prediction = pd.DataFrame({

    "Actual": y_test.values,

    "LinearRegression": pred_lr,

    "RandomForest": pred_rf,

    "XGBoost": pred_xgb

})

prediction.to_csv(
    "model_predictions.csv",
    index=False
)

print("\nPrediction File Saved Successfully.")

print("\nFirst 5 Predictions")

print(prediction.head())
```
